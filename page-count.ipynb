```python
import os

def count_pages_in_journals(root_directory):
    total_pages = 0
    journal_page_counts = {}

    # Duyệt qua tất cả các thư mục trong thư mục gốc
    for journal_folder in os.listdir(root_directory):
        journal_path = os.path.join(root_directory, journal_folder)
        
        # Kiểm tra xem có phải là thư mục không
        if os.path.isdir(journal_path):
            page_count = 0
            
            # Duyệt qua tất cả các file trong thư mục tạp chí
            for file_name in os.listdir(journal_path):
                if file_name.endswith('.txt'):
                    page_count += 1
            
            # Cập nhật số trang cho tạp chí
            journal_page_counts[journal_folder] = page_count
            total_pages += page_count

    return journal_page_counts, total_pages

# Đường dẫn tới thư mục gốc
root_directory = '3.txt-output-saved'
journal_page_counts, total_pages = count_pages_in_journals(root_directory)

# In kết quả ra màn hình
for journal, count in journal_page_counts.items():
    print(f"Tạp chí: {journal} - Số trang: {count}")

print(f"Tổng số trang của tất cả các tạp chí: {total_pages}")
```
