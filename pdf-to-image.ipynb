```python
import os
from pdf2image import convert_from_path

def convert_pdf_to_images(pdf_path, output_dir):
    """Chuyển đổi PDF thành danh sách các hình ảnh và lưu từng trang thành file riêng."""
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    images = convert_from_path(pdf_path)  # Chuyển đổi toàn bộ trang của PDF
    
    for page_number, image in enumerate(images):
        image_path = os.path.join(output_dir, f"page_{page_number + 1}.png")
        image.save(image_path, format='PNG')  # Lưu từng trang thành file PNG
        print(f"Đã lưu trang {page_number + 1} vào '{image_path}'.")

def convert_all_pdfs_in_folder(input_folder):
    """Chuyển đổi tất cả các file PDF trong thư mục thành hình ảnh."""
    for filename in os.listdir(input_folder):
        if filename.endswith('.pdf'):
            pdf_path = os.path.join(input_folder, filename)
            output_directory = os.path.join(input_folder, filename[:-4])  # Tạo tên thư mục từ tên file PDF
            print(f"Đang chuyển đổi '{pdf_path}'...")
            convert_pdf_to_images(pdf_path, output_directory)

if __name__ == "__main__":
    input_folder = '1.pdf-input'  # Điền đường dẫn tới thư mục chứa các file PDF

    # Chuyển đổi tất cả các PDF trong thư mục
    convert_all_pdfs_in_folder(input_folder)
```
